```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Fractional Ionization
def X(S):
    return((-1+(1+4*S)**(1/2))/(2*S))

# Where S is defined as
def S(T, n):
    k = 8.617E-5
    m = .511E6
    Q = 13.6
    return(3.84*n*((k*T)/m)**(3/2)*np.exp(Q/(k*T)))
```

```python
T = np.linspace(3000, 4500)
y = np.linspace(3000, 4500)
y.fill(.5)
n1 = 4E-10
n2 = 8E-10

plt.plot(T, X(S(T, n1)), label='$n=4E-10$')
plt.plot(T, X(S(T, n2)), label='$n=8E-10$')
plt.plot(T, y, label='$X=0.5$')
plt.plot(3721, 0.5, 'o', label='3721K')
plt.plot(3785, 0.5, 'o', label='3785K')
plt.xlabel('Temperature')
plt.ylabel('Fractional Ionization')
plt.title("Recombination")
plt.legend()
plt.show()
```

The change in $\eta$=4E-10 to $\eta$=8E-10 increases the temperature of recombination, $T_{rec}$, by roughly 64K.
